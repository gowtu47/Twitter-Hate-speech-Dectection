```python
import pandas as pd
```

```python
df=pd.read_csv(r'clean_data.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df['text']=df['text'].astype(str)
```

```python
df.dtypes
```

```python
import re
import numpy as np
df.text=df.text.str.replace("@\w*", " ") #removes twitter handles
```

```python
df.text=df.text.str.replace("[^a-zA-Z]", " ") #removes any characters other than the alphabets
```

```python
df.head()
```

```python
df.text=df.text.str.replace("#\w*",' ')
```

```python
import nltk
```

```python
from nltk.tokenize import word_tokenize
```

```python
from nltk.corpus import stopwords
```

```python
stop_word = set(stopwords.words('English'))
```

```python
df1=df.copy()
```

```python
df1=df1.apply(lambda x: x.astype(str).str.lower()) #lower-casing all the tweets
```

```python
df1['text'].str.split()
```

```python
df1['new_text']=df1['text'].apply(lambda x: ' '.join([item for item in x.split() if item not in stop_word]))
```
